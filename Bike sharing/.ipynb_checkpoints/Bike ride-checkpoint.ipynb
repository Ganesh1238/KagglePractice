```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
train.columns
```

```python
test['casual'] = 0
test['registered'] = 0
test['count'] = 0
```

```python
# removed outliers from the data

train = train[np.abs(train["count"]-train["count"].mean())<=(3*train["count"].std())] 
```

```python
print(train.shape)
print(test.shape)
```

```python
merge_data = pd.concat([train, test], ignore_index=True)
```

```python
train['windspeed'].value_counts()
```

```python
merge_data['day'] = pd.to_datetime(merge_data.datetime).dt.day
merge_data['year'] = pd.to_datetime(merge_data.datetime).dt.year
merge_data['month'] = pd.to_datetime(merge_data.datetime).dt.month
merge_data['weekday'] = pd.to_datetime(merge_data.datetime).dt.weekday
merge_data['date'] = pd.to_datetime(merge_data.datetime).dt.date
merge_data['hour'] = pd.to_datetime(merge_data.datetime).dt.hour
merge_data['year_season']=merge_data.apply(lambda x:'{}_{}'.format(str(x['year']),str(x['season'])),axis=1)
merge_data['year_month']=merge_data.apply(lambda x:'{}_{}'.format(str(x['year']),str(x['month'])),axis=1)

merge_data['windspeed']=merge_data[['year','month','hour','windspeed']].groupby(['year','month','hour']).transform(lambda x:x.replace(0,np.median([i for i in x if i>0])))
merge_data['windspeed']=pd.cut(merge_data['windspeed'],bins=[0,20,60],labels=['0','1',])
```

```python
merge_data.head()
```

```python
merge_data['windspeed'].value_counts()
```

```python
merge_data['day_type']=0
#union_data['day_type']=
merge_data['day_type'][(merge_data['holiday']==0)& (merge_data['workingday']==0)]='weekend'
merge_data['day_type'][(merge_data['holiday']==0)& (merge_data['workingday']==1)]='workingday'
merge_data['day_type'][(merge_data['holiday']==1)]='holiday'

merge_data.day_type.value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.grid_search import GridSearchCV
from sklearn import cross_validation, metrics
from sklearn.preprocessing import OneHotEncoder

train_y = np.log1p(train[['count']]+1)
undumm = merge_data[['datetime','year_month','atemp','temp', 'humidity']]
get_dumm = merge_data[['weather','workingday','hour','weekday','windspeed','day_type']]
```

```python
dumm = pd.get_dummies(get_dumm, columns=get_dumm.columns)
train_X=pd.concat([undumm[:10739],dumm[:10739]],axis=1)
test_X=pd.concat([undumm[10739:],dumm[10739:]],axis=1)
```

```python
train_X
```

```python
regr = RandomForestRegressor(n_estimators=100, oob_score=True)
regr.fit(train_X.loc[:,'year_month':], np.ravel(train_y))
```

```python
plt.figure(figsize=(100,20))
sns.barplot(x=train_X.loc[:,'year_month':].columns,y=regr.feature_importances_)
```

```python
merge_data['count'][10739:]=np.exp(regr.predict(test_X.loc[:,'year_month':]))-1
```

```python
submission=pd.DataFrame({
        "datetime": merge_data[10739:].datetime,
        "count": merge_data[10739:]['count']
    })
submission.to_csv('bike_predictions_gbm_separate_without_fe.csv', index=False)
```
